```python
# Jika kamu tidak punya module 'core', kita gunakan list standar agar bisa jalan
class SortedArray():
    def __init__(self, max_size):
        # Menggunakan list Python biasa untuk simulasi array
        self._array = [None] * max_size 
        self._max_size = max_size
        self._size = 0

    def insert(self, value): # Gunakan 'value' agar lebih standar
        if self._size >= self._max_size:
            # Gunakan f-string (tambahkan f di depan kutip) agar variabel bisa terbaca
            raise ValueError(f"The array already full. Max size = {self._max_size}")
        
        for i in range(self._size, 0, -1):
            if self._array[i-1] <= value:
                self._array[i] = value
                self._size += 1
                return  # <--- PENTING: Harus keluar fungsi setelah data masuk!
            else:
                self._array[i] = self._array[i-1]
        
        # Bagian ini hanya jalan jika value adalah yang TERKECIL
        self._array[0] = value
        self._size += 1

    def display(self):
        print(self._array[:self._size])

    def delete(self, target):
        index = self.search(target)
        if index is None:
            raise ValueError(f'Unable to delete element {target}: the entry is not in the array')
        for i in range(index, self._size - 1):
            self._array[i] = self._array[i + 1]
        self._size 

# --- Eksekusi ---
arr = SortedArray(5)
arr.insert(30)
arr.insert(10)
arr.insert(20)
arr.insert(15)

print("Isi array sekarang:")
arr.display() # Output seharusnya: [10, 15, 20, 30]

arr.delete(20)
arr.display
```
