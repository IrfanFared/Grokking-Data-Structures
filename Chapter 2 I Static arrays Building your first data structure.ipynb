```python
#EXERCISES
#2.1â€‚ Write the code for a function returning the minimum value in an array and its index.
#Hint: Can you adapt the function max_in_array?

#solution 

array_dummy = [42,23,22,1,4,2]

def min_in_array(array):
    if len(array) == 0:
        raise Exception("there is no array at all")
    min_index =0
    for i in range(1, len(array)):
        if array[i] < array[min_index]:
            min_index = i
    return min_index,array[min_index]

min_in_array(array_dummy)

```
