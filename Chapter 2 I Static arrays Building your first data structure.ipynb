```python
#EXERCISES
#2.1  Write the code for a function returning the minimum value in an array and its index.
#Hint: Can you adapt the function max_in_array?

#solution 

array_dummy = [42,23,22,1,4,2]

def min_in_array(array):
    if len(array) == 0:
        raise Exception("there is no array at all")
    min_index =0
    for i in range(1, len(array)):
        if array[i] < array[min_index]:
            min_index = i
    return min_index,array[min_index]

min_in_array(array_dummy)

```

```python
#2.2  Can you write a method returning both the max and min values at once? What’s the
#advantage of computing both within the same method

array_dummy = [42,23,22,1,4,2]

def min_in_array(array):
    if len(array) == 0:
        raise Exception("there is no array at all")
    min_index =0
    max_index = 0
    for i in range(1, len(array)):
        if array[i] < array[min_index]:
            min_index = i
        elif array[i] > array[max_index]:
            max_index = i
    return min_index,array[min_index],"|",max_index,array[max_index]

min_in_array(array_dummy)
```

```python
import random

class LinearSearch:
    def __init__(self, data):
        self.data = data

    def linear_search(self, target):
        for i, value in enumerate(self.data):
            if value == target:
                print(f"Target ditemukan pada indeks ke-{i}")
                return i
        print("Target tidak ditemukan")
        return None


data = [random.randint(1, 100) for _ in range(10)]
print("Data:", data)

ls = LinearSearch(data)
result = ls.linear_search(10)
print("Hasil return:", result)
```
