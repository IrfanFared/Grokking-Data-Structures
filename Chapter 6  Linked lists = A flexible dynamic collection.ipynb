```python
# Apa itu linked list?
# linked list adalah sebuah strutur data berantai yang terhubung dengan node,yang mana ketika kita menghasese sebuah linked list kita harus tidak bisa jump ke indek yang kita tentukan tapi harus terurut .Yang akibat koplesiketas aksesnya liniar



# Membuat Linked list

class Node:
    def __init__(self,data):
        self.data = data 
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None


#inisisasi node

node1 = Node(10)
node2 = Node(30)
node3 = Node(20)

# proses menghubungkan linked list

llk = LinkedList()
llk.head = node1
node1.next = node2
node2.next = node3


#contoh penampilan linked list

temp = llk.head

while temp:
    print(temp.data, end="---->")
    temp = temp.next
print("none")
```
