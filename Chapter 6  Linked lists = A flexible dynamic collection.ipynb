```python
# Apa itu linked list?
# linked list adalah sebuah strutur data berantai yang terhubung dengan node,yang mana ketika kita menghasese sebuah linked list kita harus tidak bisa jump ke indek yang kita tentukan tapi harus terurut .Yang akibat koplesiketas aksesnya liniar



# Membuat Linked list

class Node:
    def __init__(self,data):
        self.data = data 
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None


#inisisasi node

node1 = Node(10)
node2 = Node(30)
node3 = Node(20)

# proses menghubungkan linked list

llk = LinkedList()
llk.head = node1
node1.next = node2
node2.next = node3


#contoh penampilan linked list

temp = llk.head

while temp:
    print(temp.data, end="---->")
    temp = temp.next
print("none")
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    # Fungsi untuk menambah data di akhir (Append)
    def tambah_data(self, data_baru):
        new_node = Node(data_baru)
        if self.head is None:
            self.head = new_node
            return
        
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    # Fungsi Search (Pencarian)
    def cari_nilai(self, x):
        current = self.head
        indeks = 0
        
        # Mulai menelusuri dari head sampai ketemu None
        while current is not None:
            if current.data == x:
                return f"Ketemu! Nilai {x} ada di urutan ke-{indeks}"
            
            current = current.next # Pindah ke node berikutnya
            indeks += 1
            
        return f"Yah, nilai {x} tidak ada di dalam list."

    # Fungsi untuk menampilkan seluruh isi list
    def cetak_list(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

# --- Cara Penggunaan ---
irfan_list = LinkedList()

# 1. Masukkan data
irfan_list.tambah_data("Python")
irfan_list.tambah_data("Sains Data")
irfan_list.tambah_data("Machine Learning")

# 2. Cetak list untuk melihat strukturnya
print("Isi Linked List:")
irfan_list.cetak_list()

# 3. Coba Cari Data
print("\n--- Testing Pencarian ---")
print(irfan_list.cari_nilai("Sains Data"))
print(irfan_list.cari_nilai("Java"))
```
